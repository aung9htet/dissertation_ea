```python
from algorithm_to_test.symmetric_mexpohd_multiprocessing import immune_algorithm
import numpy as np
import sys
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
#starts from 5 bits until the max_bit
#repeat is the number of time repeated and the average time taken from it will be used
def plot_graph_onemax(max_bit, repeat):
    y_axis_one = np.array([])
    y_axis_five = np.array([])
    y_axis_ten = np.array([])
    y_axis_fifteen = np.array([])
    y_axis_twenty = np.array([])
    control_y = np.array([])
    sys.stdout.write('The process will go through ' + str(max_bit) + ' bits')
    sys.stdout.flush()
    for n in range(5,max_bit+1):
        control_value = np.log(n) * n
        control_y = np.append(control_y, control_value)
        run_time = 0
        for i in range(repeat):
            run_time += immune_algorithm(n, 1, 0)
        run_time = run_time/repeat
        y_axis_one = np.append(y_axis_one, run_time)
        run_time = 0
        for i in range(repeat):
            run_time += immune_algorithm(n, 5, 0)
        run_time = run_time/repeat
        y_axis_five = np.append(y_axis_five, run_time)
        run_time = 0
        for i in range(repeat):
            run_time += immune_algorithm(n, 10, 0)
        run_time = run_time/repeat
        y_axis_ten = np.append(y_axis_ten, run_time)
        run_time = 0
        for i in range(repeat):
            run_time += immune_algorithm(n, 15, 0)
        run_time = run_time/repeat
        y_axis_fifteen = np.append(y_axis_fifteen, run_time)
        run_time = 0
        for i in range(repeat):
            run_time += immune_algorithm(n, 20, 0)
        run_time = run_time/repeat
        y_axis_twenty = np.append(y_axis_twenty, run_time)
        #To check progress
        sys.stdout.write(f"\r{' '*100}\r")
        sys.stdout.flush()
        sys.stdout.write('Currently working on ' + '\x1b[7;37;42m' + str(n) + 'th' + '\x1b[0m' + ' bit out of ' + str(max_bit) + ' bits')
        sys.stdout.flush()
    x_axis = np.arange(5,max_bit+1)
    plt.xlabel("number of bits")
    plt.ylabel("run time by number of evaluations")
    plt.plot(x_axis,y_axis_one, label = "Symmetric MexpoHD T=1")
    plt.plot(x_axis,y_axis_five, label = "Symmetric MexpoHD T=5")
    plt.plot(x_axis,y_axis_ten, label = "Symmetric MexpoHD T=10")
    plt.plot(x_axis,y_axis_fifteen, label = "Symmetric MexpoHD T=15")
    plt.plot(x_axis,y_axis_twenty, label = "Symmetric MexpoHD T=20")
    plt.plot(x_axis,control_y, label = "nlogn")
    plt.legend()
    
plot_graph_onemax(30, 20)
```
